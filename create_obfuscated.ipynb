```python
import pandas as pd
import numpy as np
import math
import os
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
imgs = np.load('mnist_files/mnist_img.npy')
for i in range(10, 71, 5):
    masks = np.load('mask/mask_' + str(i) + '.npy')
    proc_img = np.where(masks == 0, imgs, 0) 
    np.save('mnist_obfuscated/mnist_' + str(i), proc_img)
    # Display the first image in the processed_imgs array
    plt.imshow(proc_img[0], cmap='gray')  # Use 'gray' for grayscale images
    plt.title('First Processed Image')
    plt.axis('off')  # Turn off the axis
    plt.show()
```
